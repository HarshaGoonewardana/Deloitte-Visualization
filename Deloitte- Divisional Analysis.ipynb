```python
# Imports 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python

df=pd.read_pickle('time_diff.pkl')
```

```python
df.sample(5)
```

```python
# Pace Distribution
```

```python
df['Division'].value_counts().sort_index()
```

```python
df[df['Gender']!='male']['Division'].value_counts().sort_index()
```

```python
ax = sns.barplot(x="Division", y="Pace", hue="Gender", data=df)
plt.rcParams['figure.figsize'] = (8.75, 7.0);
plt.title ('Pace distribution by division')
plt.ylabel('Pace (min/mile)');
```

```python
ax = sns.barplot(x="Division", y="n_total_time", hue="Gender", data=df,palette="bright")
plt.rcParams['figure.figsize'] = (8.75, 7.0);
plt.title ('Net time distribution by division');
```

```python
 ax = sns.barplot(x="Division", y="g_total_time", hue="Gender", data=df, estimator=np.min)
plt.title ('Best Gun time distribution by division')
plt.ylabel('Gun time (minutes)');
```

```python
ax = sns.barplot(x="Division", y="n_total_time", hue="Gender", data=df, estimator=np.min,palette="Set1")
plt.title ('Best Net time distribution by division')
plt.ylabel('Net time (minutes)');
```

```python
ax = sns.countplot(x="Division", data=df,hue='Gender',palette="nipy_spectral_r",
order = df['Division'].value_counts().index)
plt.title ('Number of participants by division')
plt.ylabel('Partcipants');
```
